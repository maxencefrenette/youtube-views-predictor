### Load Libraries

```python
import numpy as np
import pandas as pd
```

### Load YouTube data from CSV files

```python
# Read CSV file
yt_data_1 = pd.read_csv('../data/youtube_video_data_20230304.csv')

# Drop the CSV file's numeric row count/index column
yt_data_1 = yt_data_1.drop('Unnamed: 0', axis=1)

# Convert 2 date columns into datetime objects
yt_data_1['video_published_at'] = pd.to_datetime(yt_data_1['video_published_at'])
yt_data_1['channel_published_at'] = pd.to_datetime(yt_data_1['channel_published_at'])

# Output the dataframe
yt_data_1
```

```python
# Read CSV file
yt_data_2 = pd.read_csv('../data/youtube_video_data_20230403.csv')

# Drop the CSV file's numeric row count/index column
yt_data_2 = yt_data_2.drop('Unnamed: 0', axis=1)

# Convert 2 date columns into datetime objects
yt_data_2['video_published_at'] = pd.to_datetime(yt_data_2['video_published_at'])
yt_data_2['channel_published_at'] = pd.to_datetime(yt_data_2['channel_published_at'])

# Output the dataframe
yt_data_2
```

### Drop records for any videos also found in the 2nd CSV file 

Data in the first CSV file was pulled from YouTube on March 4, 2023. 
Data in the second CSV file was pulled from YouTube on April 3, 2023. 
If the 2nd file contains data for a video ID that exists in the 1st file, 
that data is more up-to-date, so we want to keep it.

```python
print("Data Frame 1 Shape Before:", yt_data_1.shape)

updated_videos = yt_data_1.loc[yt_data_1['video_id'].isin(yt_data_2['video_id'])]['video_id']
print("Found updated data on", len(updated_videos), "videos")

yt_data_1 = yt_data_1.loc[~yt_data_1['video_id'].isin(updated_videos)]

print("Data Frame 1 Shape After:", yt_data_1.shape)
```

### Now combine data from both Data Frames

```python
print("Data Frame 1 Shape Before:", yt_data_1.shape)
print("Data Frame 2 Shape Before:", yt_data_2.shape)

yt_data = pd.concat([yt_data_1, yt_data_2])

print("Combined Data Frame Shape After:", yt_data.shape)
```

### Deduplicate any records within the final Data Frame

```python
print("Data Frame Shape Before:", yt_data.shape)

yt_data.drop_duplicates(subset='video_id', keep='last', inplace=True)

print("Data Frame Shape After:", yt_data.shape)
```

### Output Data Frame to CSV file

```python
yt_data.to_csv('youtube_video_data.csv')
```
